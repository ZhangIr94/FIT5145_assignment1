## Analysis of annual income of provinces in China

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
#annual income of each province
income = pd.read_csv('annual_income.csv', index_col = 'Region')
income.index = income.index.set_names([''])
income = income.transpose()
income.index = income.index.astype(int)
income = income.reindex(sorted(income.columns), axis=1)
income = income.reindex(sorted(income.index), axis=0)
income.head()
```

```python
plt.plot(income.index,income.loc[:,'Anhui':'Zhejiang'])
plt.show()
```

### Beijing, Henan and Zhejiang income

```python
incomeBHZ = pd.DataFrame(income, columns=['Beijing','Henan','Zhejiang'])
incomeBHZ.plot.line()
plt.xlabel('Year')
plt.ylabel('annual income')
plt.xticks(np.arange(2008, 2018, step=1))
plt.show()
```

## Analysis of population growth rate in various provinces of China

```python
#population growth rate for each province
growth_rate = pd.read_csv('growth_rate.csv', index_col = 'Region')
growth_rate.index.names = ['']
growth_rate = growth_rate.transpose()
growth_rate.index = growth_rate.index.astype(int)
growth_rate = growth_rate.reindex(sorted(income.columns), axis=1)
growth_rate = growth_rate.reindex(sorted(income.index), axis=0)
growth_rate.head()
```

### Population growth rate of Bejing, Henan and Zhejiang

```python
growthRateBHZ = pd.DataFrame(growth_rate, columns=['Beijing','Henan','Zhejiang'])
growthRateBHZ.plot.line()
plt.xlabel('Year')
plt.ylabel('Population growth rate')
plt.xticks(np.arange(2008, 2018, step=1))
plt.show()
```

## Analysis of the higher education population in China's provinces

```python
#Higher education population per 100,000 people for each province
higher_education_population = pd.read_csv('higher_education_population.csv', index_col = 'Region')
higher_education_population.index.names = ['']
higher_education_population = higher_education_population.transpose()
higher_education_population.index = higher_education_population.index.astype(int)
higher_education_population = higher_education_population.reindex(sorted(higher_education_population.columns), axis=1)
higher_education_population = higher_education_population.reindex(sorted(higher_education_population.index), axis=0)
higher_education_population.head()
```

```python
higherEduBHZ = pd.DataFrame(higher_education_population, columns=['Beijing','Henan','Zhejiang'])
higherEduBHZ.plot.line()
plt.xlabel('Year')
plt.ylabel('Higher education population')
plt.xticks(np.arange(2008, 2018, step=1))
plt.show()
```

### Higher education population prediction

```python
higherEduB = pd.DataFrame(higher_education_population, columns=['Beijing'])
higherEduB.plot.line()
plt.xlabel('Year')
plt.ylabel('Higher education population in Beijing')
plt.xticks(np.arange(2008, 2020, step=1))

slope, intercept, r_value, p_value, std_err = linregress(higherEduB.index,higherEduB['Beijing']) # y = kx + b
line = [slope*xi + intercept for xi in higherEduB.index]
plt.plot(higherEduB.index,line,'r-', linewidth=3)
plt.scatter(higherEduB.index, higherEduB.Beijing)
plt.show()
```

```python
x2025 = slope*2025+intercept
int(x2025)
```

## Analysis of the relationship between the population growth rate of China's provinces, the higher education population, and the annual income

```python
relation_column = np.asarray(sorted(list(set(income.columns) | set(growth_rate.columns) | set(higher_education_population.columns))))
relation_index = np.asarray(sorted(list(set(income.index) | set(growth_rate.index) | set(higher_education_population.index))))

relationship_report = pd.DataFrame(columns = pd.MultiIndex.from_product([relation_column,['income','growth_rate','higher_edu']]),
                   index = relation_index)

for i in relation_column:    
    for j in relation_index:
        
        if ((j in income.index) & (i in income.columns)):
            relationship_report.at[j, pd.IndexSlice[ i , 'income']] = income.at[j,i]
        else:
            relationship_report.at[j, pd.IndexSlice[ i , 'income']] = np.nan
        
        if ((j in growth_rate.index) & (i in growth_rate.columns)):
            relationship_report.at[j, pd.IndexSlice[ i , 'growth_rate']] = growth_rate.at[j,i]
        else:
            relationship_report.at[j, pd.IndexSlice[ i , 'growth_rate']] = np.nan
        
        if ((j in higher_education_population.index) & (i in higher_education_population.columns)):
            relationship_report.at[j, pd.IndexSlice[ i , 'higher_edu']] = higher_education_population.at[j,i]
        else:
            relationship_report.at[j, pd.IndexSlice[ i , 'higher_edu']] = np.nan
            

relationship_report
```

### Trend of income and higher education population

```python
plt.plot(relationship_report.loc[:,pd.IndexSlice['Anhui':'Zhejiang', 'income']],relationship_report.loc[:,pd.IndexSlice['Anhui':'Zhejiang', 'higher_edu']])
plt.title('Relationship between income and higher_education_population')
plt.ylabel('higher_education_population')
plt.xlabel('income')
plt.show()
```

## Visualising the Relationship over Time

```python
from motionchart.motionchart import MotionChart
incomeData = pd.read_csv("annual_income.csv", index_col = 'Region')
incomeData.index = incomeData.index.set_names([''])
incomeData = incomeData.transpose()
incomeData.index = incomeData.index.astype(int)
incomeData = incomeData.reset_index()
incomeData.rename(columns = {'index':'Year'}, inplace = True)
incomeData = incomeData.reindex(sorted(incomeData.columns), axis=1)
income = incomeData.reindex(sorted(incomeData.index), axis=0)
incomeData = pd.melt(incomeData, id_vars='Year', var_name='Province', value_name='income')

growthData = pd.read_csv("growth_rate.csv", index_col = 'Region')
growthData.index = growthData.index.set_names([''])
growthData = growthData.transpose()
growthData.index = growthData.index.astype(int)
growthData = growthData.reset_index()
growthData.rename(columns = {'index':'Year'}, inplace = True)
growthData = growthData.reindex(sorted(income.columns), axis=1)
growthData = growthData.reindex(sorted(income.index), axis=0)
growthData = pd.melt(growthData, id_vars='Year', var_name='Province', value_name='growth_rate')

higher_eduData = pd.read_csv("higher_education_population.csv", index_col = 'Region')
higher_eduData.index = higher_eduData.index.set_names([''])
higher_eduData = higher_eduData.transpose()
higher_eduData.index = higher_eduData.index.astype(int)
higher_eduData = higher_eduData.reset_index()
higher_eduData.rename(columns = {'index':'Year'}, inplace = True)
higher_eduData = higher_eduData.reindex(sorted(income.columns), axis=1)
higher_eduData = higher_eduData.reindex(sorted(income.index), axis=0)
higher_eduData = pd.melt(higher_eduData, id_vars='Year', var_name='Province', value_name='higher_edu')

temp_df = pd.merge(incomeData,growthData, on=['Year','Province'], how='outer')
final_df = pd.merge(temp_df, higher_eduData, on=['Year', 'Province'], how='outer')

final_df
# visualiza data on motion chart
finalmChart = MotionChart(df = final_df, key='Year', x='income', y='growth_rate', xscale='linear', yscale='linear',
                     size='higher_edu', color='Province', category='Province')
finalmChart.to_notebook()
```
